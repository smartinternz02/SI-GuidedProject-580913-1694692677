```python
#import necessary libraries
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('car_crashes')
df
```

### Line Plot

```python
plt.figure(figsize=(8, 4))
sns.lineplot(data=df, x="total", y="speeding")
plt.title("Line Plot: Total Car_Crashes vs. Speeding")
plt.xlabel("Total Car_Crashes")
plt.ylabel("Speeding")
plt.show()
```

<b>Inference:</b> According to the graph, we can say that there is no straight linear relationship between the 
    car_crashes and speeding since the data points are scattered. Hence these two variables are not strongly co-related.

```python
plt.figure(figsize=(8, 4))
sns.lineplot(data=df, x="total", y="alcohol")
plt.title("Line Plot: Total Car_Crashes vs. Alcohol")
plt.xlabel("Total Car_Crashes")
plt.ylabel("Alcohol")
plt.show()
```

<b>Inference:</b> According to the graph, we can say that with increase in alcohol consumption, number of car crashes
    also increases. Eventhough some data points are scattered, most of the points satisfy the above condition.

### Bar Chart

```python
plt.figure(figsize=(15, 11))
sns.barplot(data=df, x="abbrev", y="total")
plt.title("Bar Chart: Total Crashes by State")
plt.xlabel("State Abbreviation")
plt.ylabel("Total Crashes")
plt.show()
```

<b>Inference:</b> According to the Bar chart, we say that the State "SC" has the highest total number of crashes, while State 
    "WV" has the lowest total number of crashes among the states in the dataset.

### Histogram

```python
sns.histplot(data=df, x="total", bins=10, kde=True)
plt.title("Histogram: Distribution of Total Car_Crashes")
plt.xlabel("Total Car_Crashes")
plt.ylabel("Frequency")
plt.show()
```

<b>Inference:</b> According to the histogram, we can say that most states have a relatively low total number of crashes, 
    with a few states having significantly higher crash counts. The distribution is right-skewed, with a long tail on the 
    right side.

### Scatter Plot

```python
sns.scatterplot(data=df, x="total", y="speeding")
plt.title("Scatter Plot: Total Car_Crashes vs. Speeding")
plt.xlabel("Total Car_Crashes")
plt.ylabel("Speeding")
plt.show()
```

<b>Inference:</b> According to the Scatter Plot, we can say that there is a positive correlation between the carcrashes and 
    speeding. This means with increase in speed, car crashes also increases.

### Box Plot

```python
plt.figure(figsize=(8, 4))
sns.boxplot(data=df, x="alcohol", y="total")
plt.title("Box Plot: Total Car_Crashes by Alcohol Level")
plt.xlabel("Alcohol Level")
plt.ylabel("Total Car_Crashes")
plt.show()
```

<b>Inference:</b> From the box plot, we can say that increase in alcohol levels will lead to increase in total car crashes

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x="total")
plt.title("Box Plot: Distribution of Total Car_Crashes")
plt.xlabel("Total Crashes")
plt.show()
```

<b>Inference:</b> According to the box plot, we can say that most states in the dataset have relatively low total crash counts, 
    with a few outliers having notably higher crash counts.

### Joint Plot

```python
sns.jointplot(x="total",y="ins_premium",data=df)
```

<b>Inference:</b> According to the joint plot above, we can say that there is no strong linear relationship between the 
    total car crashes and insurance premium since the data points are scattered. 

### Count Plot

```python
plt.figure(figsize=(17, 11))
sns.countplot(x="total",data=df)
```

<b>Inference:</b> The above count plot indicates the number of times different "total" value occurs.

### Dist plot

```python
sns.distplot(df['alcohol'], kde=True)
plt.title('Distribution of Alcohol Levels in Car Crashes')
plt.xlabel('Alcohol Levels')
plt.ylabel('Density')
plt.show()
```

<b>Inference:</b> The above Dist plots will show the distribution of alcohol levels in car crashers. We can get information
    about whether alcohol levels are concentrated in a specific range or if they are widely distributed.

### RelPlot

```python
sns.relplot(x='speeding', y='alcohol', data=df, kind='scatter')
plt.title('Relationship between Speeding and Alcohol Levels in Car Crashes')
plt.xlabel('Speeding')
plt.ylabel('Alcohol Levels')
plt.show()
```

<b>Inference:</b> The above Relplot will show the relationship between speeding and alcohol levels in car crashes. You can 
    observe whether there is a correlation between these two variables. If points cluster around a line or exhibit a clear
    pattern, it indicates a relationship.

### Swarmplot

```python
sns.swarmplot(x ="not_distracted", y ="total",data = df)
```

<b>Inference:</b>The swarm plot displays how "total" values are distributed within each "not_distracted" category, 
    revealing potential patterns, outliers, and differences across categories.

### Pointplot

```python
plt.figure(figsize=(16, 11))
sns.pointplot(x = "total", y = "speeding", data = df)
```

<b>Inference:</b> The point plot compares the average "total" values across different "speeding" categories, revealing 
    trends and variations in "total" while indicating the direction and significance of the relationship between the variables.

### Violin Plot

```python
plt.figure(figsize=(6, 4))
sns.violinplot(data=df, x="total")
plt.title("Violin Plot: Distribution of Total Crashes")
plt.xlabel("Total Crashes")
plt.show()
```

<b>Inference:</b> The violin plot shows that, most states have a low number of crashes but some states have a high number of 
    crashes.

### Heat Map

```python
corr  = df.corr()
corr
```

```python
sns.heatmap(corr,annot=True,cmap="YlGnBu")
```

<b>Inference:</b> The above heatmap represents the Pearson correlation coefficients which can be thought of as the 
    explainability between two arrays.
